```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import string
import re
```

```python
# Ensure required NLTK resources are available
nltk.download('punkt')
nltk.download('stopwords')

# Load data
with open("apple.txt", "r", encoding="utf-8") as file:
    text = file.read()
text
```

```python
len(text)
```

#### Text preprocessing steps

```python
# 1. Lowercase
text = text.lower()

# 2. Remove punctuation
text = text.translate(str.maketrans("", "", string.punctuation))

# 3. Remove numbers
text = re.sub(r'\d+', '', text)

# 4. Tokenize
tokens = word_tokenize(text)

tokens
```

```python
# 5. Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

# Optional: Join back for WordCloud
cleaned_text = " ".join(filtered_tokens)

# 6. Generate Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='black').generate(cleaned_text)

# 7. Display Word Cloud
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud for Apple Product Reviews", fontsize=20)
plt.show()
```
